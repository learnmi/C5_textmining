# Bayesian Inference

lets assume, 2000 users out of which 400 clapped.

```python
import numpy as np
np.setprintoptions(threshold=100)
```

# Sample Data

```python
# Set clap proba to 20
clap_proba = 0.5
clap_data = np.random.binomial(n=1, p=clap_proba, size=2000)
```

# Modelling P(Theta)

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = 400
b = 2000 - a

theta_range = np.linspace(0,1,1000)
# Prior P(theta)
prior = stats.beta.pdf(x=theta_range, a=a,b=b)
```

```python
plt.rcParams['figure.figsize'] = [20, 7]
fig, ax = plt.subplots()
plt.plot(theta_range, prior, linewidth=3, color='palegreen')
# Add a title
plt.title('[Prior] PDF of "Probability of Claps"', fontsize=20)
# Add X and y Label
plt.xlabel('Î¸', fontsize=16)
plt.ylabel('Density', fontsize=16)
# Add a grid
plt.grid(alpha=.4, linestyle='--')
```

# Plotting likelihood - P(x given theta)

```python
X = np.arange(0, len(clap_data)+1)
likelihood = stats.binom.pmf(k=X,n=len(clap_data),p=clap_proba)
```

```python
# Create the plot
fig, ax = plt.subplots()
# Add a title
plt.title('[Likelihood] Probability of Claps' , fontsize=20)
# Add X and y Label
plt.xlabel("X", fontsize=16)
plt.ylabel("Probability", fontsize=16)
# Add a grid
plt.grid(alpha=.4, linestyle='--')
plt.plot(X, likelihood, linewidth=3, color='red')
```
